# Analytical Linear Regression

```python
x = np.linspace(0, 10, 50)
y = np.linspace(0, 40, 50)

for i in range(len(y)):
    y[i] = y[i] + np.random.normal(y[i], 1, 1)
    x[i] = x[i] + np.random.normal(x[i], 0.5, 1)
```

```python
plt.plot(x, y, '.')
```

```python
X = np.matrix(x)
X = X.T

Y = np.matrix(y)
Y = Y.T

X = np.c_[np.ones(X.shape[0]),X]
```

```python
_lambda = 3

_w = (X.T @ X - _lambda*np.identity(X.shape[1])).I @ X.T @ Y
w  = (X.T @ X).I @ X.T @ Y
```

```python
def lin_reg(x):
    return float(w[0]) + float(w[1])*x

def _lin_reg(x):
    return float(_w[0]) + float(_w[1])*x
```

```python
y_fit  = list()
_y_fit = list()
x_fit  = list()

for x_ in X[:,1]:
    x_fit.append(float(x_))
    y_fit.append(lin_reg(x_fit[-1]))
    _y_fit.append(_lin_reg(x_fit[-1]))
```

```python
plt.plot(x, y, '.')

plt.plot(x_fit, y_fit, '-')
plt.plot(x_fit, _y_fit, '-')
```
